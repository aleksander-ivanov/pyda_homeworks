### 2. Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки (если покупка была, сам файл visit_log.csv изменять не надо). Запишите в файл funnel.csv визиты из файла visit_log.csv, в которых были покупки с указанием категории.
Учтите условия на данные:
содержимое purchase_log.txt помещается в оперативную память компьютера
содержимое visit_log.csv - нет; используйте только построчную обработку этого файла

```python
import json

purchases_file_path = 'data/purchase_log.txt'
visits_file_path = 'data/visit_log.csv'
result_file_path = 'data/funnel.csv'
```

```python
def purchases_to_dict(purchases):
    result = {}
    for line in purchases:
        dict_from_line = json.loads(line)

        key = dict_from_line['user_id']
        value = dict_from_line['category']

        if key == 'user_id':
            continue

        result[key] = value
        
    return result
```

```python
with open(purchases_file_path, encoding='utf-8') as f_purchases:
    purchases = f_purchases.readlines()
    purchases_dict = purchases_to_dict(purchases)

    with open(visits_file_path, encoding='utf-8') as f_visits:
            with open(result_file_path, 'w') as f_result:

                for i, line in enumerate(f_visits):
                    line = line.strip()
                    if i == 0:
                        line += ',category\n' # append column header
                        f_result.write(line)
                        print(line)

                    else:
                        user_id = line.strip().split(',')[0]
                        if user_id in purchases_dict:
                            line += ',' + purchases_dict[user_id] + '\n'
                            f_result.write(line)
                            print(line)

                            # if i > 5:  # For test 
                            #     break
```
