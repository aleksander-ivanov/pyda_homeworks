# Домашнее задание к лекции «Случайные события. Случайные величины»

Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш (см. справа)

Выгодна ли игра игроку?

Сколько в среднем приобретает
или теряет игрок за одну игру?

777: 200 руб.
999: 100 руб.
555: 50 руб.
333: 15 руб.
111: 10 руб.
*77: 5 руб.
**7: 3 руб.
*00: 2 руб.
**0: 1 руб.

– любая цифра
Дополнительное задание повышенной сложности. Теоретически рассчитать средний выигрыш (проигрыш) и сравнить с результатами моделирования

```python
import numpy as np
```

```python
trials = 1000000    # Количество попыток
successes = 0    # Количество успехов
gains = 0    # Выигрыш

for i in range(trials):
    x = np.random.randint(0, 1000)
    print(f'Выпало {x}')    
    if x == 777:
        successes += 1
        gains += 200
    elif x == 999:
        successes += 1
        gains += 100
    elif x == 555:
        successes += 1
        gains += 50
    elif x == 333:
        successes += 1
        gains += 15
    elif x == 111:
        successes += 1
        gains += 10
    elif x % 100 == 77:
        successes += 1
        gains += 5
    elif x % 10 == 7:
        successes += 1
        gains += 3
    elif x % 100 == 0:
        successes += 1
        gains += 2
    elif x % 10 == 0:
        successes += 1
        gains += 1
```

```python
# Вероятность выпада любой выиграшной комбинации
successes / trials             
```

```python
# Теоретическая вероятность
p = 1 / 1000 * 200 + 1 / 1000 * 100 + 1 / 1000 * 50 + 1 / 1000 * 15 + 1 / 1000 * 10 + 10 / 1000 * 5 + 100 / 1000 * 3 + 10 / 1000 * 2 + 100 / 1000 * 1
p
```

```python
# Средний выигрыш:
p_monte_carlo = gains/trials
p_monte_carlo
```

```python
print(f'1. Потрачено: {trials} руб. Выиграно: {gains} руб. Игра {"выгодна" if gains > trials else "не выгодна"}')
print(f'2. Игрок в среднем теряет за игру {(1 - p_monte_carlo):.2f} руб.')
print(f'3. Теоретический средний выигрыш {p:.2f} руб.')
print(f'4. Опытный средний выигрыш {p_monte_carlo:.2f} руб.')
```
