# Домашнее задание к лекции «Случайные события. Случайные величины»

Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш (см. справа)

Выгодна ли игра игроку?

Сколько в среднем приобретает
или теряет игрок за одну игру?

777: 200 руб.
999: 100 руб.
555: 50 руб.
333: 15 руб.
111: 10 руб.
*77: 5 руб.
**7: 3 руб.
*00: 2 руб.
**0: 1 руб.

– любая цифра
Дополнительное задание повышенной сложности. Теоретически рассчитать средний выигрыш (проигрыш) и сравнить с результатами моделирования

```python
import numpy as np
```

```python
def play():
    # np.random.seed(5)

    min = 0
    max = 999
    count = 1
    digits = [str(np.random.randint(min, max)) for i in range(count)]
    result = [(len(str(max)) - len(digit)) * '0' + digit for digit in digits]
    return result[0]
```

```python
trials = 1000    # Количество попыток
successes = 0    # Количество успехов
prize = 0    # Выигрыш

for i in range(trials):
    random_score = play()
    print(f'Выпало {random_score}')    
    if random_score == '777':
        successes += 1
        prize += 200
    elif random_score == '999':
        successes += 1
        prize += 100
    elif random_score == '555':
        successes += 1
        prize += 50
    elif random_score == '333':
        successes += 1
        prize += 15
    elif random_score == '111':
        successes += 1
        prize += 10
    elif random_score[-2:] == '77':
        successes += 1
        prize += 5
    elif random_score[-1:] == '7':
        successes += 1
        prize += 3
    elif random_score[-2:] == '00':
        successes += 1
        prize += 2
    elif random_score[-1:] == '0':
        successes += 1
        prize += 1
```

```python
p_monte_carlo = successes / trials    # Вероятность выпада любой выиграшной комбинации
p = 225 / 1000                        # Теоретическая вероятность, всего 225 выиграшных комбинаций
```

```python
print(f'Потрачено: {trials} руб. Выиграно: {prize} руб.')
print(f'Выигрыш с {trials} попыток составил {prize} руб.')
print(f'Вероятность из моделирования - {p_monte_carlo}')
print(f'Теоретическая вероятность - {p}')
print(f'Вероятность выигрыша - {p * 100}%')

print(f'Теоретический выигрыш с {trials} попыток = {p * trials} руб.')
print(f'Теоретический проигрыш с {trials} попыток = {trials - (p * trials)} руб.')
```
